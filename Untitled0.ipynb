TV 뿐 아니라 거실의 조명 밝기를 바꾸고 에어컨의 온도와 바람 세기를 조절할 수 있는 인공지능 홈보드도 존재합니다.
이 기능은 2019년 출시한 인공지능 TV 제품부터 제공합니다.

##**자율주행자동차**

**자율주행 자동차**는 *운전자가 차량을 조작하지 않아도 스스로 도로의 상황을 파악하여 자동으로 움직이는 자동차* 입니다. 
복잡한 도로와 교통 체계, 예기치 못한 돌발 상황들을 실시간으로 인지하여 스스로 판단, 제어하는 시스템을 기반으로 자율 주행을 합니다.

자율 주행 자동차의 기본원리는 인지, 판단, 제어로 크게 3단게로 구분이 됩니다.
1. **인지**: 자율 주행에 있어 가장 중요한 기술로 GPS와 카메라, 레이더 등을 활용해 주변 상황의 정보를 인식, 수집합니다. 빠르고 정확하게 정보를 인식하여 적절한 판단과 제어를 할 수 있는 단계입니다.
2. **판단**: 인지 단계에서 얻은 정보를 바탕으로 주행 환경과 목표 지점에 적합한 주행 전략을  수립하고 판단하여 결정하는 단계입니다.
3. **제어**: 자동차 스스로 엔진 구동과 주행 방향을 조절하고 결정하여 본격적으로 안전 운전 주행을 시작해 사고까지 예방하는 단계입니다.

국제자동차기술자협회에 따르면 자율 주행의 기술 수준은 총 6단계로, 0단계에서 5단계로 나누어져 구분됩니다.
* **0단계** - 자율 주행 기술이 없는 완전 수동 단계로 운전자가 전적으로 모든 모니터링과 조작을 제어합니다.
* **1단계** - 특정 기능 자동으로 운전자의 편의를 위한 속도와 제동에 관련하여 일부 특정 자동화 기능이 추가된 단계입니다.
* **2단계** - 상황에 맞는 속도 조절과 장애물 회피 등으로 속도와 방향을 제어합니다. 두 가지 이상 자동 주행이 가능합니다.
* **3단계** - 자율 주행 기술의 상용화를 위해 테스트가 진행되는 단계로, 운전자의 도움 없이도 교통신호와 도로 흐름까지 인식해 스스로 운전을 하는 수준입니다. 거의 모든 기능을 제어하고 특수한 경우에만 운전자가 개입합니다.
* **4단계** - 고도 자율주행으로 주변 환경에 관계없이 자동차가 스스로 모든 기능을 제어합니다. 운전자가 목적지만 설정하면 되는 단계로 스스로 안전한 자율주행이 가능합니다.
* **5단계** - 사람의 개입이 전혀 필요 없는 완벽한 자율 주행 기술 단계로, 자율 주행뿐 아니라 인공지능 비서, 건강 체크, 자동차와의 대화까지 가능합니다. 사람이 타지 않고도 주행이 가능해 무인 자동차에 가깝습니다.

하지만 자율 주행을 통해 이루어지는 자동차 사고가 빈번하게 발생하고 있습니다. 잘못을 한 사람이 운전자인지, 소프트웨어 개발사인지, 자동차 제조사 인지 확실한 가해자의 기준이 부정확하기에 정확한 규제가 필요한 상황입니다.
아직까지는 자율 주행 자동차의 상용화된 기술 단계는 0~2단계로 *사고가 났을 경우 책임은 **운전자**에게 있습니다.* 

오는 7월부터 자율주행차의 출시 및 판매가 가능하게 됐습니다. 
자율주행차 시대가 임박하자 여러 보험사들도 관련 상품 개발에 나섰습니다.

머신러닝과 딥러닝은 데이터를 의미있는 것으로 변환하는 것인데 데이터를 표현하는 방법, 데이터가 입력이 되어 출력이 되는 중의 과정, 딥러닝과 머신러닝에서 사용하는 파이썬, 파이썬을 사용하기 위한 코랩의 사용법 등 지금 강의에서 수업하는 것처럼 기초부터 꼼꼼히, 꾸준히 공부하는 것이 중요하다 생각합니다.
